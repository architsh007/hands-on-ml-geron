# K-Means Clustering — Algorithm Intuition

In this notebook we explore how the K-Means clustering algorithm works.

Goals:

• Understand the K-Means objective function  
• Visualize cluster assignments  
• Observe centroid updates  
• Understand convergence behavior  
• Introduce the Elbow Method for selecting k

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(
    n_samples=500, 
    centers=4,
    cluster_std=1.2,
    random_state=42
)

# Plot of Dataset

plt.scatter(X[:, 0], X[:, 1], s=30, alpha=0.7)
plt.title("Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

We generate a synthetic dataset using `make_blobs`.

The dataset contains naturally separable clusters, making it ideal for visualizing K-Means behavior.

```python
kmeans = KMeans(
    n_clusters=4,
    init="k-means++",
    n_init=10,
    random_state=42
)

kmeans.fit(X)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Visualize Clusters and Centroids

plt.scatter(X[:, 0], X[:, 1], c=labels, s=30, alpha=0.7, cmap="viridis")
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", s=200, marker="X", alpha=0.7)
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## Interpretation

K-Means partitions the dataset into k clusters by minimizing:

WCSS = Σ ||x - μ||²

Each point is assigned to the nearest centroid, and centroids are updated as the mean of their assigned points.

The algorithm iterates between:

1. Assignment step  
2. Centroid update step

until convergence.

```python
kmeans.inertia_
```

`inertia_` is the **Within Cluster Sum of Squares (WCSS)**.

It measures the compactness of clusters. Lower values indicate tighter clusters.

```python
k_values = range(1, 10)
wcss = []

for k in k_values:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    wcss.append(model.inertia_)
```

```python
plt.plot(k_values, wcss, marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal k")
plt.show()
```

## Elbow Method

As the number of clusters increases, WCSS decreases.

However, the improvement eventually slows down.

The optimal number of clusters is often chosen at the **elbow point**, where additional clusters provide diminishing returns.

## Key Takeaways

• K-Means partitions data using centroid-based clustering  
• The objective is minimizing within-cluster variance  
• Initialization matters due to local minima  
• K-Means works best with spherical clusters  
• The Elbow Method helps estimate the optimal number of clusters
