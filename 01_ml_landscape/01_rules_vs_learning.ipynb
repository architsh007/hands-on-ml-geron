# Rules vs Learning

This notebook demonstrates why rule-based systems fail in noisy environments
and why learning from data is more effective.

We generate synthetic data with noise and analyze why explicit rules struggle
to capture underlying patterns.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

X = np.linspace(0, 10, 50)
noise = np.random.randn(50) * 2
y = 2 * X + noise
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Data Generated from an Underlying Pattern with Noise")
plt.show()
```

## Why Rules Fail

The relationship between `X` and `y` follows a clear trend, but noise makes it
impossible to write exact rules that work for every point.

Instead of encoding rules manually, machine learning attempts to **estimate
the underlying relationship** from data.


### Try This
- Increase the noise magnitude
- Increase the number of data points
- Change the underlying relationship to nonlinear

Observe how writing explicit rules becomes harder.


```python
# Changing the noise and see the plot of data again

for noise_level in [0.5, 5, 10]:
    noise = np.random.randn(50) * noise_level
    y = 2 * X + noise

    plt.figure(figsize=(5, 3))
    plt.scatter(X, y)
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title(f"Data with Noise Level: {noise_level}")
    plt.show()
```

## Relationship Fails between `X` and `Y`
We can see from the above plots that with increased magnitude noise the relationship between `X` and `Y` is recognizable by human eye


```python
# Increasing the number of data points for plot

for num_points in [20, 100, 500]:
    X = np.linspace(0, 10, num_points)
    noise = np.random.randn(num_points) * 2
    y = 2 * X + noise

    plt.figure(figsize=(5, 3))
    plt.scatter(X, y, alpha=0.6)
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title(f"Data with Number of Points: {num_points}")
    plt.show()
```

## Relationship between `X` and `Y` as number of datapoints increased
As we increased the data points the underlying relationship between `X` and `Y` is more visible through human eye.

```python
# Changing the underlying pattern to nonlinear

X = np.linspace(0, 10, 50)
noise = np.random.randn(50) * 2
y = np.sin(X) * 5 + noise

plt.figure(figsize=(8, 5))
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Data Generated from a Nonlinear Pattern with Noise")
plt.show()
```
